```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
```

```python

USERNAME = 'test_7_account_3'
PASSWORD = '-8NqBvrVK.pf_ri'
PATH = "D:\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"
```

```python
class InstagramBot:
  def __init__(self,path):
    self.driver = webdriver.Chrome(service=Service(path))
    self.action = ActionChains(self.driver)


  def login(self):
    self.driver.get('https://www.instagram.com/')
    time.sleep(5)

    username = self.driver.find_element(By.NAME, 'username')
    password = self.driver.find_element(By.NAME, 'password')


    username.send_keys(USERNAME)
    password.send_keys(PASSWORD)
    
    time.sleep(2)
    password.send_keys(Keys.ENTER)

  def find_user(self,ACC_NAME):
   time.sleep(3)
   self.driver.get(f'https://www.instagram.com/{ACC_NAME}')
   time.sleep(3)
    
   follower = self.driver.find_element(By.CSS_SELECTOR, 'li div a span span .x5n08af x1s688f')
   self.action.click(follower)
   self.action.perform() 
```

```python
bot = InstagramBot(PATH)
bot.login()
bot.find_user('robertdowneyjr')
```
